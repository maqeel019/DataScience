```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
Eng = {"01":70 , "02":80 , "03":50 , "04":90 , "05":60 , "06":80 , "07":50 , "08":70}
pd.Series(Eng)  
Phy  = {"01":80 , "02":60 , "03":90 , "04":70 , "05":70 , "06":50 , "07":78 , "08":80}
pd.Series(Phy) 
Che  = {"01":90 , "02":70 , "03":60 , "04":80, "05":80 , "06":86 , "07":67 , "08":67}
pd.Series(Che) 
Math  = {"01":70 , "02":90 , "03":70 , "04":60, "05":90 , "06":78 , "07":87 , "08":78}
pd.Series(Math)  
Data = pd.DataFrame({"Eng" : Eng, "Phy": Phy , "Math" : Math , "Che":Che })
print(Data)
fig, ax = plt.subplots()

# Plotting the data
chart = ax.plot(Che.keys(), Data['Che'] , '*-m' , ms=10 ,  mec = 'k' ,mfc = 'm' ,label="Che") #mec = markedEdgeColor   linestyle = ls = dotted : /dashed -
chart = ax.plot(Che.keys(), Data['Phy'], 'p-m' ,ms=10 ,  mec = 'r' ,mfc = 'g' , label="Phy")   #mfc = markedfaceColor   linewidth = lw = 20
chart = ax.plot(Che.keys(), Data['Eng'], 'h-m' , ms=10 ,  mec = 'b' ,mfc = 'r',label="Eng")   #'*-m' = marker|line|color / can use hexaXolors or 
chart = ax.plot(Che.keys(), Data['Math'], 'v-m' , ms=10 ,  mec = 'g' ,mfc = 'b',label="Math") #ms = markedSize



# above60 = Data[Data['Eng'] > 60]
# print(above60)
# ax.plot(above60)
# Set x-axis tick positions and labels
ax.set_xticklabels(Math.keys())

# Adding labels and title 
font1 = {'family':'serif','color':'Red','size':15}
font2 = {'family':'serif','color':'Green','size':15}

ax.set_xlabel('RollNO' , fontdict = font1)
ax.set_ylabel('Marks' , fontdict = font2)
ax.set_title('Marks of Student',  loc = 'left')
#Specify Which Grid Lines to Display : axis = 'x' , axis = 'y' , default valus is both
# can set for grid : color = 'green', linestyle = '--', linewidth = 0.5
# ax.grid(axis = 'x')
# Adding legend
ax.legend()
# Display the plot
# plt.show()
# chart
```
